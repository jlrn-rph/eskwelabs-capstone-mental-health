```python
#import libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/engagement_by_flair.csv')
```

```python
df
```

```python
index = ['num_comments', 'score', 'upvote_ratio']
df = df.rename(index={0: index[0], 1: index[1], 2: index[2]})

```

```python
df2 = df.T.reset_index()
df2
```

```python
df2.to_csv('csv/flair_EDA.csv', index=False)
```

```python
df_comments = df2[['index','num_comments']].sort_values('num_comments', ascending=False)
df_comments.to_csv('csv/flair_comments.csv', index=False)
```

```python
df_score = df2[['index','score']].sort_values('score', ascending=False)
df_score.to_csv('csv/flair_score.csv', index=False)
```
